```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Load Dataset
df = pd.read_csv('UCI_Credit_Card.csv')
```

```python
# 1. Data Exploration
print("\nFirst 5 rows of the dataset:\n")
print(df.head())

print("\nBasic statistics:\n")
print(df.describe())

print("\nMissing values per column:\n")
print(df.isnull().sum())

print("\nData types of each column:\n")
print(df.dtypes)
```

```python
# Visualize Age Distribution
plt.figure(figsize=(6, 2))
sns.histplot(df['AGE'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Credit Card Clients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize Gender Distribution
plt.figure(figsize=(6, 2))
sns.countplot(x='SEX', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender (1=Male, 2=Female)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize Education Level
plt.figure(figsize=(8, 4))
sns.countplot(x='EDUCATION', data=df, palette='muted')
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize Credit Limit Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['LIMIT_BAL'], bins=30, kde=True, color='coral')
plt.title('Credit Limit Distribution')
plt.xlabel('Credit Limit')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Customer Segmentation using K-Means
X = df[['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1']]
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LIMIT_BAL', y='BILL_AMT1', hue='Cluster', data=df, palette='viridis')
plt.title('Customer Segmentation based on Balance and Bill Amount')
plt.legend(title='Cluster')
plt.show()
```

```python
# PCA for Dimensionality Reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='bright')
plt.title('Customer Segmentation using PCA')
plt.legend(title='Cluster')
plt.show()
```
